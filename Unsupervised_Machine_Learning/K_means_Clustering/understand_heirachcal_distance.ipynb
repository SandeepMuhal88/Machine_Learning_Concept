```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


datasets=[(1,2),(2,3),(3,4),(8,7),(8,8),]
x,y=zip(*datasets)
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='blue',s=100,label='Data Points')
for i, point in enumerate(datasets):
    plt.text(point[0], point[1], f' {i+1}', fontsize=12)
    
```

```python
# Perform Herachical Clustering
x=np.array([[1, 2], [2, 3], [3, 4], [8, 7], [8, 8], [25, 80]])
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

linked=linkage(x,method='ward')
```

```python
plt.figure(figsize=(10,7))
dendrogram(linked,orientation='top',distance_sort='ascending',show_leaf_counts='True')
plt.show()
```

```python

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
lables=cluster.fit_predict(x)
print("cluster lable",lables)
```
