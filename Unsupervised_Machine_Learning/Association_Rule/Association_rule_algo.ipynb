```python
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.datasets import fetch_openml
mashroom_data=fetch_openml(name='mushroom', version=1, as_frame=True)
df=mashroom_data.data
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
df_encoded=pd.get_dummies(df)
df_encoded.head()
```

```python
freq_itemsets=apriori(df_encoded, min_support=0.3, use_colnames=True)
```

```python
rules=association_rules(freq_itemsets, metric='confidence', min_threshold=1.5)
rules.head()
```
