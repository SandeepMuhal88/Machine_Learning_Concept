Isolation Forest Anomaly Detection

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.ensemble import IsolationForest

# Step 1: Generate Synthetic Data (Clusters + Some Outliers)
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Add some outliers (random points far from clusters)
np.random.seed(42)
outliers = np.random.uniform(low=-10, high=10, size=(10, 2))  # 10 outliers
X = np.vstack([X, outliers])  # Combine normal data and outliers

# Step 2: Train Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # 5% contamination (outliers)
labels = iso_forest.fit_predict(X)

# Step 3: Visualize Results
plt.figure(figsize=(8, 6))

# Plot normal points (label = 1)
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], c='green', label='Normal Points', s=50)

# Plot anomalies (label = -1)
plt.scatter(X[labels == -1, 0], X[labels == -1, 1], c='red', label='Anomalies (Outliers)', marker='x', s=100)

plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
print("Anomaly Labels:", labels)
```
